CENIPA - Ocorrências Aeronáuticas na Aviação Civil Brasileira
Dados de ocorrências aeronáuticas da aviação civil brasileira nos últimos 10 anos.
link:https://dados.gov.br/dataset/ocorrencias-aeronauticas-da-aviacao-civil-brasileira

OCORRÊNCIA.csv - Informações sobre as ocorrências.
OCORRÊNCIA_TIPO.csv - Informações sobre o tipo de ocorrência.
AERONAVE.csv - Informações sobre as aeronaves envolvidas nas ocorrências.
FATOR_CONTRIBUINTE.csv - Informações sobre os fatores contribuinte das ocorrências que tiveram investigações finalizadas.

```python
#importando bilbiotecas
import pandas as pd
import matplotlib.pyplot as plt


#carregando arquivos
ocorrencias = pd.read_csv('ocorrencia.csv', sep=';', usecols=['ocorrencia_dia','codigo_ocorrencia','codigo_ocorrencia1','codigo_ocorrencia3', 'ocorrencia_classificacao','ocorrencia_cidade','ocorrencia_uf','ocorrencia_aerodromo', 'ocorrencia_hora','total_aeronaves_envolvidas'])
ocorrencias_tipo = pd.read_csv('ocorrencia_tipo.csv', sep=';', usecols=['codigo_ocorrencia1','ocorrencia_tipo'])
fator_contrib = pd.read_csv('fator_contribuinte.csv', sep=';', usecols=['codigo_ocorrencia3','fator_nome', 'fator_area'])

#Formatanto datas e criando colunas Qtd
ocorrencias['data_ocorrencia'] = pd.to_datetime(ocorrencias['ocorrencia_dia'],format='%d/%m/%Y')
ocorrencias['Qtd'] = 1

#concatenando dataframes 
ocorrencias = pd.merge(ocorrencias, ocorrencias_tipo, how ='inner', on = ['codigo_ocorrencia1'])
ocorrencias = pd.merge(ocorrencias, fator_contrib, how ='inner', on = ['codigo_ocorrencia3'])

#organizando frame por data
ocorrencias.sort_values(by='data_ocorrencia', inplace=True)

```

Total de ocorrências por ano

```python
ocorrencias['ano_ocorrencia'] = ocorrencias['data_ocorrencia'].dt.year 
totAno = ocorrencias.filter(items=['ano_ocorrencia', 'Qtd'])
totAno = totAno.groupby(['ano_ocorrencia']).count()
plt.figure(figsize=[16,4])
plt.fontsize=(13)
plt.ylabel('Quantidade')
plt.xlabel('Anos')
plt.plot(totAno, color='red', linestyle='dashed', linewidth=4)
plt.title('Evolução Ocorrências x Ano (2010-2020)\n')
plt.show()
```

Top 10: Estados x Ocorrencias 

```python
totUf = ocorrencias.filter(items=['ocorrencia_uf', 'Qtd'])
totUf = totUf.groupby(['ocorrencia_uf']).count().sort_values('Qtd', ascending=False).head(10)
totUf.plot.bar(
    figsize=[16,4],
    fontsize=13,
    title='Top 10: Estados x Ocorrências (2010-2020)\n')
```

Top 10 cidades X ocorencias

```python
totCidade = ocorrencias.groupby(['ocorrencia_cidade']).size().sort_values(ascending=False)
totCidade.head(10).plot.pie(
        title='Top 10: Ocorrências x UF (2010-2020)\n',
        figsize=(16,10), 
        fontsize=13,
        autopct='%1.1f%%'
 
    )
```

Total Ocorrencias por dia da semana

```python
ocorrencias['dia_ocorrencia'] = ocorrencias['data_ocorrencia'].dt.day_name()

totDia = ocorrencias.groupby(['dia_ocorrencia']).count()


totDia = ocorrencias.filter(items=['dia_ocorrencia', 'Qtd'])
totDia= totDia.groupby(['dia_ocorrencia']).count().sort_values('Qtd', ascending=False)

totDia.plot.bar(
    figsize=[16,4],
    fontsize=13,
    title='Ocorrẽncias x Dia da Semana (2010-2020)\n')
```

Top 10 tipo de Ocorrencias

```python
totTipoOcorrencia = ocorrencias.filter(items=['ocorrencia_tipo', 'Qtd'])
totTipoOcorrencia = totTipoOcorrencia.groupby(['ocorrencia_tipo']).count().sort_values('Qtd',ascending=False).head(10)
totTipoOcorrencia.plot.line(
    color='blue',
    linestyle='dashed',
    linewidth=4,
    subplots=True,
    figsize=[16,4],
    fontsize=10,
    title='Top 10: Ocorrências x Tipo (2010-2020)\n')
```

Top 3 Ocorrencias

```python
totFatorAreaa = ocorrencias.filter(items=['fator_area', 'Qtd'])
totFatorAreaa= totFatorAreaa.groupby(['fator_area']).count().sort_values('Qtd', ascending=False).head(3)

totFatorAreaa.plot.bar(
    figsize=[16,4],
    fontsize=13,
    title='Top 3: Ocorrẽncia x Fator Contribuinte (2010-2020)\n')
```
